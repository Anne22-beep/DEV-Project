## I decided to use excel to do more data wrangling and came up with the below dataset

```python
df = pd.read_excel('Solar.xlsx')
df.head(25)
```

```python
df.info()
```

```python
months = df['Month'].unique()
months

dictSolar = {i:round(df[df['Month']==i]['Solar electricity generation (kWh)'].mean(),2) for i in months}
dictSolar
```

```python
plt.figure(figsize=(12,8))
plt.title('Average Monthly Sola Electricity Generation')
sns.barplot(dictSolar,color='g')
plt.xlabel('Month')
plt.ylabel('Solar Generation in kWh')

plt.show()
```

```python
dictElectricity = {i:round(df[df['Month']==i]['Electricity usage (kWh)'].mean(),2) for i in months}
dictElectricity
```

```python
plt.figure(figsize=(12,8))
plt.title('Average Monthly Electricity Consumption')
sns.barplot(dictElectricity,color='y',legend='auto')
plt.xlabel('Month')
plt.ylabel('Consumption in  kWh')

plt.show()
```

```python
days = df['Day'].unique()
days
```

```python
dictDaySolar = {i:round(df[df['Day']==i]['Solar electricity generation (kWh)'].mean(),2) for i in days}
dictDayElectricity = {i:round(df[df['Day']==i]['Electricity usage (kWh)'].mean(),2) for i in days}
dictDayElectricity
```

```python
scale = np.linspace(0,2.5,10,endpoint=False)
scale
```

```python
plt.figure(figsize=(12,8))
plt.title('Average Daily Solar Generation in kWh')
sns.lineplot(dictDaySolar,color='y')
sns.lineplot(dictDayElectricity,color='b')
plt.xlabel('Day')
plt.ylabel('Generation in  kWh')
plt.yticks(scale)
plt.legend(['Solar Generation','Electricity Usage'])

plt.show()
```
